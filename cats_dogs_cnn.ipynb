```python
import cv2
import numpy as np
import os
from random import shuffle
from tqdm import tqdm
import matplotlib.pyplot as plt
from keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint, Callback, EarlyStopping
```

```python
# CONSTANTS
TRAIN_DIR = './data_cnn/train/'
TEST_DIR = './data_cnn/test/'
N_ROWS = 64
N_COLS = 64
N_CHANNELS = 3
```

```python
train_images = [TRAIN_DIR + item for item in os.listdir(TRAIN_DIR)]
test_images = [TEST_DIR + item for item in os.listdir(TEST_DIR)]
# shuffle train images
shuffle(train_images)
```

**read & preprocess image**

```python
# helper functions
def read_image(file_path):
    img = cv2.imread(file_path,cv2.IMREAD_COLOR)
    return(img)
def preprocess_image(img):
    # resize
    img = cv2.resize(img,(N_ROWS,N_COLS))
    # rescale
    img = img / 255
    return(img)
```

```python
img = read_image(train_images[2])
img = preprocess_image(img)
plt.imshow(img)
```

**prepare data**

```python
def prepare_data(images):
    n_images = len(images)
    data = np.zeros((
        n_images,N_ROWS,N_COLS,N_CHANNELS))
    for i, file_path in enumerate(images):
        img = read_image(file_path)
        img = preprocess_image(img)
        data[i] = img
        if i%1000 == 0: print('processed {} out of {}'.format(i,n_images))
    return(data)
```

```python
data_train = prepare_data(train_images)
data_test = prepare_data(test_images)
print(data_train.shape,data_test.shape)
```

**prepare class labels: 1 if cat, 0 if dog**

```python
def prepare_labels(images):
    labels = np.array([1 if 'cat' in item else 0 for item in images]).reshape(-1,1)
    return(labels)
```

```python
labels = prepare_labels(train_images)
```

## Baseline model: LeNet

```python
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.models import Sequential
from keras.optimizers import Adam

lenet_model = Sequential()
lenet_model.add(Conv2D(6, (5, 5), activation='relu',name='conv1',
                        input_shape=(N_ROWS,N_COLS,3)))
lenet_model.add(MaxPooling2D((2, 2), name='pool1'))
lenet_model.add(Conv2D(16, (5, 5), activation='relu', name='conv2'))
lenet_model.add(MaxPooling2D((2, 2), name='pool2'))
lenet_model.add(Flatten(name='flatten'))
lenet_model.add(Dense(120, activation='relu', name='fc1'))
lenet_model.add(Dense(84, activation='relu', name='fc2'))
lenet_model.add(Dense(1, activation='sigmoid', name='predictions'))
# compile lenet model
lenet_model.compile(loss='binary_crossentropy',
              optimizer=Adam(),
              metrics=['accuracy'])
```

```python
lenet_model.summary()
```

**fit lenet model**

```python
batch_size = 50
epochs = 10
history = lenet_model.fit(data_train, labels, batch_size=batch_size, epochs=epochs,
              validation_split=0.2, verbose=1, shuffle=True)
```

**plot loss function and accuracy**

```python
def plot_perf(history):
    acc = history.history['acc']
    val_acc = history.history['val_acc']
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    epochs = range(len(acc))

    plt.plot(epochs, acc, 'bo', label='Training acc')
    plt.plot(epochs, val_acc, 'b', label='Validation acc')
    plt.title('Training and validation accuracy')
    plt.legend()

    plt.figure()

    plt.plot(epochs, loss, 'bo', label='Training loss')
    plt.plot(epochs, val_loss, 'b', label='Validation loss')
    plt.title('Training and validation loss')
    plt.legend()

    plt.show()
# plot loss and accuracy
plot_perf(history)
```

## LeNet - add Dropout layer

```python
from keras.layers import Dropout
lenet_model2 = Sequential()
lenet_model2.add(Conv2D(6, (5, 5), activation='relu',name='conv1',
                        input_shape=(N_ROWS, N_COLS, 3)))
lenet_model2.add(MaxPooling2D((2, 2), name='pool1'))
lenet_model2.add(Conv2D(16, (5, 5), activation='relu', name='conv2'))
lenet_model2.add(MaxPooling2D((2, 2), name='pool2'))
lenet_model2.add(Flatten(name='flatten'))
# The new dropout layer
lenet_model2.add(Dropout(0.5))
lenet_model2.add(Dense(120, activation='relu', name='fc1'))
lenet_model2.add(Dense(84, activation='relu', name='fc2'))
lenet_model2.add(Dense(1, activation='sigmoid', name='predictions'))

lenet_model2.compile(loss='binary_crossentropy',
              optimizer=Adam(),
              metrics=['accuracy'])
```

```python
lenet_model2.summary()
```

**plot LeNet model with Dropout**

```python
history = lenet_model2.fit(data_train, labels, batch_size=batch_size, epochs=epochs,
              validation_split=0.2, verbose=1, shuffle=True)
```

```python
# plot loss and accuracy
plot_perf(history)
```

**use pretrained VGG16 model**

```python
from keras.applications import VGG16
conv_base = VGG16(weights='imagenet',
                  include_top=False,
                  input_shape=(N_ROWS,N_COLS, 3))
```

```python
vgg_model = Sequential()
vgg_model.add(conv_base)
vgg_model.add(Flatten())
vgg_model.add(Dense(256, activation='relu'))
vgg_model.add(Dense(1, activation='sigmoid'))
```

```python
vgg_model.summary()
```

```python
# freeze network
conv_base.trainable = False
```

```python
vgg_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = vgg_model.fit(data_train, labels, batch_size=batch_size, epochs=epochs,
              validation_split=0.2, verbose=1, shuffle=True)
```

```python
# plot loss and accuracy
plot_perf(history)
```
