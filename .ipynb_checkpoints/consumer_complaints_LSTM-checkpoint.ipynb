## In this notebook we will apply LSTM to classify product categories using consumer complains in raw text format

```python
import numpy as np
import pandas as pd
import re
from nltk.corpus import stopwords
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, Dense, SpatialDropout1D, LSTM
from sklearn.model_selection import train_test_split
from keras.callbacks import EarlyStopping
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('consumer_complaints.csv')
print(df.shape)
df.head()
```

**11 product categories:**

```python
df['product'].value_counts()
```

**drop rows with missing values in columns product, consumer_complaint_narrative**

```python
df.dropna(subset=['product','consumer_complaint_narrative'],inplace=True)
# shape after dropping missing values
df.shape
```

**example of complain**

```python
df['consumer_complaint_narrative'][:1].values[0]
```

```python
text = df['consumer_complaint_narrative'][:1].values[0]
```

## Text preprocessing
- remove "X", remove digits
- convert text to lower case
- remove punctuation
- remove stop words

```python
def text_preprocess(text):
    # remove "X"
    text = text.replace('X','')
    # remove digits
    text = re.sub('[\d+]','',text)
    # convert to lower case
    text = text.lower()
    # remove punctuation
    text = re.sub('[^\w\s]', '', text)
    # remove stop words
    stop = stopwords.words('english')
    text = " ".join(text for text in text.split() if text not in stop)
    return(text)
```

```python
df['consumer_complaint_narrative'] = df['consumer_complaint_narrative'].apply(text_preprocess)
df.head()
```

## Tokenize
- apply keras tokenizer
- notice that words that have rare occurrance are mostly misspelings/mistakes

```python
# max number of words
MAX_WORDS = 20000
# max number of words in each complaint
MAX_SEQUENCE_LENGTH = 200
EMBEDDING_DIM = 100
tokenizer = Tokenizer(num_words=MAX_WORDS)
tokenizer.fit_on_texts(df['consumer_complaint_narrative'].values)
word_index = tokenizer.word_index
print('number of unique tokens {}'.format(len(word_index)))
print(word_index)
```

**transform texts into sequence of integers**

```python
X = tokenizer.texts_to_sequences(df['consumer_complaint_narrative'].values)
X = pad_sequences(X,maxlen=MAX_SEQUENCE_LENGTH)
X.shape
```

**dummify target column**

```python
Y = pd.get_dummies(df['product']).values
Y.shape
```

**train test split**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.01, random_state = 1)
print(X_train.shape,Y_train.shape)
print(X_test.shape,Y_test.shape)
```

## Fit LSTM model
- First layer performs embedding
- LSTM layer has 100 memory units
- Output layer has 11 values (correponding to 11 classes)
- Activation function is softmax, loss function is categorical_crossentropy as we have multi-class classification

```python
model = Sequential()
model.add(Embedding(MAX_WORDS, EMBEDDING_DIM, input_length=X.shape[1]))
model.add(SpatialDropout1D(0.2))
model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(11, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

epochs = 5
batch_size = 64
# callbacks=[EarlyStopping(monitor='val_loss', patience=3, min_delta=0.0001)]
history = model.fit(X_train, Y_train, epochs=epochs, batch_size=batch_size,validation_split=0.1)
```

**evaluate model on test set**

```python
accuracy = model.evaluate(X_test,Y_test)
print('Test set\n  Loss: {:0.3f}\n  Accuracy: {:0.3f}'.format(accuracy[0],accuracy[1]))
```

```python
plt.title('Loss')
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.title('Accuracy')
plt.plot(history.history['acc'], label='train')
plt.plot(history.history['val_acc'], label='test')
plt.legend()
plt.grid(True)
plt.show();
```
